```python
import itertools
```

```python
movie_list = []
movie_list_unique = []
#reading movie data form movies.txt--------------------------------------------
with open("movies.txt", "r") as m_id:
    for row in m_id:
        str_line = row.strip().split(';')
        movie_list.append(str_line)
        
        for single_movie in str_line:
            #separate unique movies in a separete list
            if(single_movie not in movie_list_unique):
                movie_list_unique.append(single_movie)

print(len(movie_list))
# movie_list
```

```python
#Helper functions for apriori algorithm=========================================================================================
# This function takes dic and minimum support as parameters and 
# eliminate a movie which has lower frequency than minimum support
def eliminate_minsup(dic, minsup):
    words = dic.keys()
    for word in list(words):
        if dic[word] < minsup:
            del dic[word]
    
    return dic


# Counting the movies in the first step
def count_first_freq(inputs, minsup):
    dic = dict()
    for line in inputs:
        for word in line:
            if word not in dic.keys():
                dic[word] = 1
            else:
                dic[word] += 1
    
    return eliminate_minsup(dic, minsup)


# Sort movie dictionary by frequency
def sorting(dic):
    sorted_dic = dict()
    alpha_sort = sorted(list(dic.keys()))
    num_sort = sorted(alpha_sort, key=dic.get, reverse=True)
    for i in num_sort:
        sorted_dic[i] = dic[i]
    
    return sorted_dic
```

```python
#calculating movies supports--------------------------------------------------------------------------------------------------- 
# minimum_support = 0.05
minimum_support = 444

frequent_single_movies = count_first_freq(movie_list, minimum_support)
print(len(frequent_single_movies))

#sorting the dictionary decending in order
frequent_single_movies = sorting(frequent_single_movies)   
print(frequent_single_movies)

# save data into oneItems.txt text file
f = open("oneItems.txt", "w")
for key, value in frequent_single_movies.items(): 
        f.write('%s:%s\n' % (value, key))
f.close()
```

```python
#Helper functions for apriori algorithm=========================================================================================
def sublist(lst1, lst2):
    return set(lst1) <= set(lst2)

def check_subset_frequency(itemset, l, key_item_length):
    if key_item_length>1:    
        subsets = list(itertools.combinations(itemset, key_item_length))
    else:
        subsets = itemset
    for iter1 in subsets:
        if not iter1 in l:
            return False
    return True

# Create string dictionary from tuple to count its frequency
def make_string_dic(dic):
    str_dic = dict()
    for key in list(dic.keys()):
        if type(key) == type(tuple()):
            string = ""
            for item in key:
                string += item + "; "
            string = string[:-1]
            str_dic[string] = dic[key]
        else:
            str_dic[key] = dic[key]
            
    return str_dic

def run_apriori(pre_candidate, records, minimum_support_count, key_item_length):
    pre_candidate = list(pre_candidate.keys())
    if key_item_length > 2:
        pre_candidate = sorted(pre_candidate)
        
    pre_candidate_comb = []
    if key_item_length > 2:
        pre_candidate_comb = sorted(list(set([item for t in pre_candidate for item in t])))
        pre_candidate_comb = list(itertools.combinations(pre_candidate_comb, key_item_length))
    else:
        pre_candidate_comb = list(itertools.combinations(pre_candidate, key_item_length))
    
    c3 = {}
    updated_candidate = {}
    for iter1 in pre_candidate_comb:
        count = 0
        for iter2 in records:
            if sublist(iter1, iter2):
                count+=1
        c3[iter1] = count
    for key, value in c3.items():
        if value >= minimum_support_count:
            if check_subset_frequency(key, pre_candidate, key_item_length-1):
                updated_candidate[key] = value 
        
    return updated_candidate
```

```python
key_item_length = 2
all_candidate = []
par_candidate = frequent_single_movies
while True:
    candidate = run_apriori(par_candidate, 
                                       movie_list, 
                                       minimum_support_count, 
                                       key_item_length)
    par_candidate = candidate
    key_item_length += 1
    
    if len(candidate) < 1:
        break
        
    all_candidate.append(candidate)


all_candidate
```

```python

all_frequent_itemset = {}
for item in all_candidate:
    all_frequent_itemset.update(item)

all_frequent_itemset_str = make_string_dic(all_frequent_itemset)
# keys = sorted(string_dic, key=lambda l: (-string_dic[l], l))

for key in all_frequent_itemset_str:
    print(all_frequent_itemset_str[key], ":", "[", key, "]", sep="")
print("")

# save data into text file
f = open("patterns.txt", "w")
for key in all_frequent_itemset_str: 
        f.write('%s: %s\n' % (all_frequent_itemset_str[key], key))
f.close()
```
